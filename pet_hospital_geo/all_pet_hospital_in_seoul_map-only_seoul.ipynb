#### 서울 범위를 벗어나는 곳들 좌표조절해서 수정

```python
url = 'https://maps.googleapis.com/maps/api/geocode/json'

for name in hospitals:
    params = {'sensor':'false', 'address':name }
    try:
        r = requests.get(url, params=params)
        
        print(name + '좌표 받기 시작...')
        
        results = r.json()['results']
        
        for result in results:
            latlng = result['geometry']['location']
            if (latlng['lng'] >126.7683)and(latlng['lng']<127.1799)and(latlng['lat']>37.4308)and(latlng['lat']<37.699):
                location = result['geometry']['location']

        print(name + '좌표 저장 완료...')
        
        small_list =list( [ location['lat'], location['lng'] ] )
        hos_dicts.update({name:small_list})
        
    except:
        print(name + '병원이 지도에 없습니다.')
```
